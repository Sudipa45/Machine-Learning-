```python
#romove column no 0 as it is not containing any meaningfull information
df=df.iloc[:,1:]
#check dataset wheather column remove or not
df.head(3)
```

```python
#import matplotlib to plot graph for visualizing data
import matplotlib.pyplot as plt
#visualize
plt.scatter(df['cgpa'],df['iq'],c=df['placement'])
```

```python
# split the dataset into features (X) and the target variable (y).
x=df.iloc[:,0:2]#':' means select all rows and '0:2' select column 0 & 1
y=df.iloc[:,-1]#select only the last column
x
```

```python
y
```

```python
#import train test split for split the dataset
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2)# pass variable to train_test_split function & save splited value
```

```python
#StandardScaler: Used to standardize features by removing the mean and scaling to unit variance.
#fit_transform: Combines fitting and transforming in one step, calculating the necessary statistics and applying the transformation to the data.
#Purpose: Standardization ensures features have similar scales, which can improve the performance and convergence of machine learning algorithms

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()#create object of SandardScaler
X_train=scaler.fit_transform(X_train)
X_test=scaler.fit_transform(X_test)


```

```python
X_train
```

```python
X_test
```

```python
#model selection
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()#crete object of model
classifier.fit(X_train,y_train)#fit training data to selected model
```

```python
#C: Regularization strength (inverse).
#class_weight: Class weights for handling imbalanced datasets.
#dual: Use dual or primal formulation.
#fit_intercept: Whether to include an intercept.
#intercept_scaling: Scaling factor for the intercept term.
#l1_ratio: Mix between L1 and L2 regularization (for elasticnet).
#max_iter: Maximum iterations for solver convergence.
#multi_class: Strategy for multi-class classification.
#n_jobs: Number of CPU cores to use.
#penalty: Type of regularization (l2).
#random_state: Seed for randomness control.
#solver: Optimization algorithm (lbfgs).
#tol: Convergence tolerance.
#verbose: Verbosity level of the solver.
#warm_start: Reuse previous solution for initialization.
#These parameters allow you to customize the logistic regression model according to the needs of your specific problem, ensuring better performance and control over the learning process.

LogisticRegression(C=1.0, class_weight=None, dual=False, fit_intercept=True,
                   intercept_scaling=1, l1_ratio=None, max_iter=200,
                   multi_class='auto', n_jobs=None, penalty='l2',
                   random_state=None, solver='lbfgs', tol=0.0001, verbose=0,
                   warm_start=False)
```

```python
y_pred=classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
#see decision boundary
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X_train, y_train.values, clf=classifier, legend=2)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Decision Regions')
```

```python
#Save the model to a file
import pickle
pickle.dump(classifier,open('model.pkl','wb'))
```

```python
# Load the model from the file
with open('model.pkl', 'rb') as file:
    loaded_clf = pickle.load(file)
```

```python
# Make predictions with the loaded model
predictions = loaded_clf.predict(X_test)
print(predictions)
```

# Pickling: Serializing the object and writing it to a file.
# Unpickling: Reading the serialized byte stream from the file and converting it back into a Python object.
